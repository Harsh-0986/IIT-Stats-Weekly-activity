## Experiment: Tossing an unbiased coin
- 1000 times
- 5000 times

## Event: Probability of getting a head


### Here we use the following notations:
- 1 for heads
- 0 for tails

## Computations

### Manual computation
Since the coin is unbiased
$$
P(\text{ Head }) = P(\text{ Tail }) = 0.5
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Latex

# Defining a function to stimulate uniform distribution
def uniform(n, m):
  '''
  Accepts:
    n: Number of outcomes in sample space
    m: m outcomes selected randomly from 1 to n

    Return:
      ndarray of type ints
  '''
  return np.random.randint(0, n, size = m)
```

# Estimating probability by simulation - Monte Carlo

The probability of an event $A$ can be estimated as follows. We can simulate the experiment repeatedly and independently, say $N$ times, and count the number of times the event occurred, say $N_A$.

A good estimate of $P(A)$ is the following:
$$P(A) \approx \frac{N_A}{N}$$
As $N$ grows larger and larger, the estimate becomes better and better. This method is generally termed as Monte Carlo simulation.

We will first evaluate probability of coin toss described above using Monte Carlo simulations. There are two steps: generate a large number of tosses and count the number of heads or tails.

In our case if we run the simulation multiple times to see what probability estimate is obtained each time then we will see that the estimate is close to 0.5.

Comparing with the above definition of Monte-Carlo simulation, we get $N=1000$ and $N_A = \text{ number of times head occured}$ and print $P(A)$ as the output

```python
# Tossing the coin 1000 times
total_heads = 0
prob_heads = []
# Counting number of 1's, i.e, heads in the output returned by function
for i in range(1000):
  heads = uniform(2, 1)
  total_heads += heads

  # Calculating prob values
  prob_value = total_heads / (i+1)

  prob_heads.append(prob_value)

# Plotting the monte carlo simulation
plt.axhline(y=0.5,color='red',linestyle='-')
plt.xlabel('iteration')
plt.ylabel('probability')
plt.plot(prob_heads)
print()
display(Latex(f"$P(A): {prob_heads[-1]}$"))
print()
```

In this case we can see that the probability is not really close to 0.5 but if we increase the range from 1000 to 5000 we get close to 0.5 from the definition of Monte-Carlo simulation.

Let's check this stament out in the following cell
